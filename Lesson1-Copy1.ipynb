```python
# Darstellen eines Bildes im Jupiter Notebook (Nicht Bestandteil der Unterrichtseinheit)
import cv2   # Bibliothek muss hinzugefügt werden "pip install opencv-python"

import matplotlib.pyplot as plt
import matplotlib

img = cv2.imread("Sprachen.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
def show_image(img, pixel_size_in_inch=0.02):
    fig_height_inch = int(img.shape[0] * pixel_size_in_inch)
    fig_width_inch  = int(img.shape[1] * pixel_size_in_inch)
    plt.figure(figsize=(fig_height_inch,fig_width_inch))
    plt.imshow(img)
    
show_image(img)
print("Source: https://spectrum.ieee.org/static/interactive-the-top-programming-languages-2018")
```
